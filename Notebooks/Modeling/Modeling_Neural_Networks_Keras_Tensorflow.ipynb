## 4. Features Selection: Define/Select Features<a id='4_Features_Selection:_Define/Select_Features'></a>


```python
# let's check our final df columns:
df.columns
```

```python
# Let's define/select our features:
features = ['property_type', 'zipcode', 'beds', 'baths', 'sqft','lot_size', 
            'days_on_market', 'price_per_sqft', 'hoa_month','property_age', 
            'school_rating', 'school_count', 'median_income','hotness_score', 
            'supply_score', 'demand_score', 'university_count','barts', 'malls', 
            'universities', 'bedBath', 'AvgRoomSize']

# And, Now let's define X and y:
X = df[features]
y = df.price
```

## 5. Train/Test Split<a id='5_Train/Test_Split'></a>

Now, let's split our data (X) into training and testing splits, without letting a model learn anything about the test split.

let's have 75/25 Train/Test split

```python
# Import train_test_split
from sklearn.model_selection import train_test_split


# Split into train and test sets:
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.25,
                                                    random_state=1981)
```

## 6. Scaling and Transformation: Standardization<a id='6_Scaling_and_Transformation:_Standardization'></a>

**Standardization** finds the mean of our data and centers the distribution around it, calculating the number of standard deviations away from the mean each point is. The number of standard deviations are then used as our new values. This centers the data around 0 but technically has no limit to the maximum and minimum values.



```python
# Import StandardScaler:
from sklearn.preprocessing import StandardScaler

# Instantiate StandardScaler:
scaler = StandardScaler()

# Fit & transform  scaler to the data:
X_train = scaler.fit_transform(X_train.astype(np.float))
X_test = scaler.transform(X_test.astype(np.float))
```

## 7. Modeling: Neural Network using keras and tensorflow<a id='7_Modeling:_Neural_Network_using_keras_and_tensorflow'></a>

**Neural networks** consist of simple input/output units called neurons (inspired by neurons of the human brain). These input/output units are interconnected and each connection has a weight associated with it. Neural networks are flexible and can be used for both classification and regression. In this article, we will see how neural networks can be applied to regression problems.

**Regression** helps in establishing a relationship between a dependent variable and one or more independent variables. Regression models work well only when the regression equation is a good fit for the data. Most regression models will not fit the data perfectly. Although neural networks are complex and computationally expensive, they are flexible and can dynamically pick the best type of regression, and if that is not enough, ***hidden layers*** can be added to improve prediction. 



### 7.1 Develop a Baseline Neural Network Model<a id='7.1_Develop_a_Baseline_Neural_Network_Model'></a>

#### 7.1.1 Create a Shallow Neural Network Architecture:<a id='7.1.1_Create_a_Shallow_Neural_Network_Architecture'></a>

We first construct the simplest neural network model with the following code, which consists of only 3 fully connected layers:
 * Input layer 1: 22 neurons, ReLU Activation.
 * Hidden layer 1: 22 neurons, ReLU Activation.
 * Output Layer: 1 neuron, Linear Activation.

```python
# First, let's Import the necessary modules from keras to create our first Neural Network:
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import Adam

# Create Model:
# Let's store our model in the variable 'model', and we'll describe it sequentially (layer by layer):
model = Sequential(name="Base_Model")

# Input layer
# We'll be using 'relu' activation and Neurons(22) and input_shape referred to the number of features:
# We have only one Hidden layer
model.add(Dense(22,activation='relu', input_shape=(22,), name="Input_Layer")) # Dense is the commonly used fully connected layer

# Hidden layer
model.add(Dense(22,activation='relu', name="Hidden_Layer_1"))

# Output layer
model.add(Dense(1, name="Output_Layer"))

# Compile Model:
model.compile(optimizer='adam',
              loss='mse',
              metrics=['mae']) # To track MAE on top of the loss funtion.
```

#### 7.1.2 Neural Network: Training and Fitting<a id='7.1.2_Neural_Network:_Training_and_Fitting'></a>

```python
# Training and fitting:
hist = model.fit(x=X_train,y=y_train,
                 validation_data=(X_test,y_test),
                 shuffle=False, verbose=1, # shuffle=False for reproducibility
                 batch_size=128,epochs=400)
```

#### 7.1.3 Neural Network: Model Summary<a id='7.1.3_Neural_Network:_Model_Summary'></a>

```python
model.summary()
```

**Param Summary:**
 * Param = (input data dimension + 1) * number of neurons. (The reason why is increased by 1 is that each neuron has a Bias).
 * Total Params = 506 (input layer) + 506 (hidden layer) + 23 (output layer).
 * Param (input/hidden layer) =( 22 + 1 ) * 22 = 506
 * Param (output layer) =( 22 + 1 ) * 1 = 23

#### 7.1.4 Neural Network: Visualizing Training and Validation Loss<a id='7.1.4_Neural_Network:_Visualizing_Training_and_Validation_Loss'></a>

```python
fig = plt.figure(figsize=(15,8))
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper right')
plt.show()
```

#### 7.1.5 Neural Network: Evaluation<a id='7.1.5_Neural_Network:_Evaluation'></a>

```python
# Predict on the test data: y_pred_nn_1
y_pred_nn_1 = model.predict(X_test)
```

```python
# Import necessary modules
from sklearn import metrics

print("********NN keras Regression Model 1 Evaluation********")
print("-------------------------------------------------")
print("NN keras Regression Model 1 - r2 Score: {:.4f}".format(metrics.r2_score(y_test, y_pred_nn_1) * 100), '%')
print("-------------------------------------------------")
print("NN keras Regression Model 1 - MAE: {:.4f}".format(metrics.mean_absolute_error(y_test, y_pred_nn_1)))
print("-------------------------------------------------")
print("NN keras Regression Model 1 - MSE:{:.4f}".format(metrics.mean_squared_error(y_test, y_pred_nn_1)))
print("-------------------------------------------------")
print("NN keras Regression Model 1 - RMSE:{:.4f}".format(np.sqrt(metrics.mean_squared_error(y_test, y_pred_nn_1))))
print("-------------------------------------------------")
print("NN keras Regression Model 1 - Variance Score:{:.4f}".format( metrics.explained_variance_score(y_test, y_pred_nn_1) * 100), '%')
print("-------------------------------------------------")
```

As we can see, the value of **root mean squared error (RMSE) is 175,471.2649**, which is slightly less than 20% of the mean value of the Sales Price i.e.  $887,769.19. also, we have a **poor variance score at 79.3446 %.** 

This mean that our base Neural Network Model (One Hidden Layer) is performing poorly compared to all other previous Regression Models.

**Next:**
 * Deeper Neural network by adding more Hidden Layers.
 * Deeper and Wider Neural network by adding more Hidden Layers and Neurons. 

#### 7.1.6 Neural Network Actual vs Prediction Visualization<a id='7.1.6_Neural_Network_Actual_vs_Prediction_Visualization'></a>

```python
# Let's plot the predictions against the actual result. Use scatter()
plt.figure(figsize=(15,10))
plt.scatter(y_test,y_pred_nn_1, color='red', marker='o', s=20, alpha=0.3,label='Actual vs Prediction-NN Model 1')
plt.title('NN keras Regression Model 1: predictions vs. the actual values')
plt.plot([0,(np.max(y_test))],[0,(np.max(y_test))], 'blue', linewidth=30, alpha=0.2)
plt.legend()
plt.xlabel('Actual Result')
plt.ylabel('Prediction')
```

### 7.2 Develop a Deeper Neural Network Model<a id='7.2_Develop_a_Deeper_Neural_Network_Model'></a>

#### 7.2.1 Create a Deeper Neural Network Architecture:<a id='7.2.1_Create_a_Deeper_Neural_Network_Architecture'></a>

Second, let's construct a deeper neural network model with the following code, which consists of only 5 fully connected layers:
 * Input layer 1: 22 neurons, ReLU Activation.
 * Hidden layer 3: 22 neurons, ReLU Activation.
 * Output Layer: 1 neuron, Linear Activation.

```python
# Create Model_2:
# Let's store our model in the variable 'model', and we'll describe it sequentially (layer by layer):
model_2 = Sequential(name="Model_3H_L") # 3H_L referred to the use of 3-Hidden layers

# Input layer
# We'll be using 'relu' activation and Neurons(22) and input_shape referred to the number of features:
# We have only one Hidden layer
model_2.add(Dense(22,activation='relu', input_shape=(22,), name="Input_Layer")) # Dense is the commonly used fully connected layer

# Hidden layer
model_2.add(Dense(22,activation='relu', name="Hidden_Layer_1"))
model_2.add(Dense(22,activation='relu', name="Hidden_Layer_2"))
model_2.add(Dense(22,activation='relu', name="Hidden_Layer_3"))

# Output layer
model_2.add(Dense(1, name="Output_Layer"))

# Compile Model:
model_2.compile(optimizer='adam',
              loss='mse',
              metrics=['mae']) # To track MAE on top of the loss funtion.
```

#### 7.2.2 Neural Network: Training and Fitting<a id='7.2.2_Neural_Network:_Training_and_Fitting'></a>

```python
# Training and fitting:
hist_2 = model_2.fit(x=X_train,y=y_train,
                 validation_data=(X_test,y_test),
                 shuffle=False, verbose=1, # shuffle=False for reproducibility                     
                 batch_size=128,epochs=400)
```

#### 7.2.3 Neural Network: Model Summary<a id='7.2.3_Neural_Network:_Model_Summary'></a>

```python
model_2.summary()
```

**Param Summary:**
 * Param = (input data dimension + 1) * number of neurons. (The reason why is increased by 1 is that each neuron has a Bias).
 * Total Params = 506 (input layer) + 506 (hidden layer) + 23 (output layer).
 * Param (input/hidden layers) =( 22 + 1 ) * 22 = 506
 * Param (output layer) =( 22 + 1 ) * 1 = 23

#### 7.2.4 Neural Network: Visualizing Training and Validation Loss<a id='7.2.4_Neural_Network:_Visualizing_Training_and_Validation_Loss'></a>

```python
fig = plt.figure(figsize=(15,8))
plt.plot(hist_2.history['loss'])
plt.plot(hist_2.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper right')
plt.show()
```

#### 7.2.5 Neural Network: Evaluation<a id='7.2.5_Neural_Network:_Evaluation'></a>

```python
# Predict on the test data: y_pred_nn_2
y_pred_nn_2 = model_2.predict(X_test)
```

```python
# Import necessary modules
from sklearn import metrics

print("********NN keras Regression Model 2 Evaluation********")
print("-------------------------------------------------")
print("NN keras Regression Model 2 - r2 Score: {:.4f}".format(metrics.r2_score(y_test, y_pred_nn_2) * 100), '%')
print("-------------------------------------------------")
print("NN keras Regression Model 2 - MAE: {:.4f}".format(metrics.mean_absolute_error(y_test, y_pred_nn_2)))
print("-------------------------------------------------")
print("NN keras Regression Model 2 - MSE:{:.4f}".format(metrics.mean_squared_error(y_test, y_pred_nn_2)))
print("-------------------------------------------------")
print("NN keras Regression Model 2 - RMSE:{:.4f}".format(np.sqrt(metrics.mean_squared_error(y_test, y_pred_nn_2))))
print("-------------------------------------------------")
print("NN keras Regression Model 2 - Variance Score:{:.4f}".format( metrics.explained_variance_score(y_test, y_pred_nn_2) * 100), '%')
print("-------------------------------------------------")
```

As we can see, the value of **root mean squared error (RMSE) is 74,143.5980**, which is slightly less than 10% of the mean value of the Sales Price i.e.  $887,769.19. also, we have a a good **variance score at 96.3320 %.** 

**This mean that our deeper Neural Network Model (3 Hidden Layers) Variance Score has improved by 21.4% from 79.3446% to 96.3320%.**

**Next:**
 * Deeper and Wider Neural network by adding more Hidden Layers and Neurons.

```python
# Let's plot the predictions against the actual result. Use scatter()
plt.figure(figsize=(15,10))
plt.scatter(y_test,y_pred_nn_1, color='black', marker='D', s=40, alpha=0.3, label='Actual vs Prediction-NN Model 1 (base)')
plt.scatter(y_test,y_pred_nn_2, color='red', marker='o', s=20, alpha=0.3,label='Actual vs Prediction-NN Model 2')
plt.title('NN keras Regression Models: predictions vs. the actual values')
plt.plot([0,(np.max(y_test))],[0,(np.max(y_test))], 'blue', linewidth=30, alpha=0.2)
plt.legend()
plt.xlabel('Actual Result')
plt.ylabel('Prediction')
```

### 7.3 Develop a Deeper and Wider Neural Network Model<a id='7.3_Develop_a_Deeper_and_Wider_Neural_Network_Model'></a>

#### 7.3.1 Create a Deeper and Wider Neural Network Architecture:<a id='7.3.1_Create_a_Deeper_and_Wider_Neural_Network_Architecture'></a>

Finally, let's construct a deeper and wider neural network model with the following code, which consists of 7 fully connected layers:
 * Input layer 1: 40 neurons, ReLU Activation (Almost Doubling Neurons from 22 to 40).
 * Hidden layer 5: 40 neurons, ReLU Activation  (Almost Doubling Neurons from 22 to 40).
 * Output Layer: 1 neuron, Linear Activation.

```python
# Create Model 3:
# Let's store our model in the variable 'model', and we'll describe it sequentially (layer by layer):
model_3 = Sequential(name="Model_5H_L_40") # 5H_L referred to the use of 5-Hidden layers.

# Input layer
# We'll be using 'relu' activation and Neurons(40) and input_shape referred to the number of features:
# We have only one Hidden layer
model_3.add(Dense(40,activation='relu', input_shape=(22,), name="Input_Layer")) # Dense is the commonly used fully connected layer

# Hidden layer
model_3.add(Dense(40,activation='relu', name="Hidden_Layer_1"))
model_3.add(Dense(40,activation='relu', name="Hidden_Layer_2"))
model_3.add(Dense(40,activation='relu', name="Hidden_Layer_3"))
model_3.add(Dense(40,activation='relu', name="Hidden_Layer_4"))
model_3.add(Dense(40,activation='relu', name="Hidden_Layer_5"))

# Output layer
model_3.add(Dense(1, name="Output_Layer"))

# Compile Model:
model_3.compile(optimizer='adam',
              loss='mse',
              metrics=['mae']) # To track MAE on top of the loss funtion.
```

#### 7.3.2 Neural Network: Training and Fitting<a id='7.3.2_Neural_Network:_Training_and_Fitting'></a>

```python
# Training and fitting:
hist_3 = model_3.fit(x=X_train,y=y_train,
                 validation_data=(X_test,y_test),
                 shuffle=False, verbose=1, # shuffle=False for reproducibility                   
                 batch_size=128,epochs=400)
```

#### 7.3.3 Neural Network: Model Summary<a id='7.3.3_Neural_Network:_Model_Summary'></a>

```python
model_3.summary()
```

**Param Summary:**
 * Param = (input data dimension + 1) * number of neurons. ***(The reason why is increased by 1 is that each neuron has a Bias).***
 * Total Params = 920 (input layer) + 1640*5 (hidden layer) + 41 (output layer) = 9,161
 * Param (input_layer) =( 22 + 1 ) * 40 = 920
 * Param (hidden layer) =( 40 + 1 ) * 40 = 1640 * 5
 * Param (output layer) =( 40 + 1 ) * 1 = 41

#### 7.3.4 Neural Network: Visualizing Training and Validation Loss<a id='7.3.4_Neural_Network:_Visualizing_Training_and_Validation_Loss'></a>

```python
fig = plt.figure(figsize=(15,8))
plt.plot(hist_3.history['loss'])
plt.plot(hist_3.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper right')
plt.show()
```

#### 7.3.5 Neural Network: Evaluation<a id='7.3.5_Neural_Network:_Evaluation'></a>

```python
# Predict on the test data: y_pred_nn_3
y_pred_nn_3 = model_3.predict(X_test)
```

```python
# Import necessary modules
from sklearn import metrics

print("********NN keras Regression Model 3 Evaluation********")
print("-------------------------------------------------")
print("NN keras Regression Model 3 - r2 Score: {:.4f}".format(metrics.r2_score(y_test, y_pred_nn_3) * 100), '%')
print("-------------------------------------------------")
print("NN keras Regression Model 3 - MAE: {:.4f}".format(metrics.mean_absolute_error(y_test, y_pred_nn_3)))
print("-------------------------------------------------")
print("NN keras Regression Model 3 - MSE:{:.4f}".format(metrics.mean_squared_error(y_test, y_pred_nn_3)))
print("-------------------------------------------------")
print("NN keras Regression Model 3 - RMSE:{:.4f}".format(np.sqrt(metrics.mean_squared_error(y_test, y_pred_nn_3))))
print("-------------------------------------------------")
print("NN keras Regression Model 3 - Variance Score:{:.4f}".format( metrics.explained_variance_score(y_test, y_pred_nn_3) * 100), '%')
print("-------------------------------------------------")
```

As we can see, the value of **root mean squared error (RMSE) is 31,832.1523**, which is slightly less than 3% of the mean value of the Sales Price i.e.  $887,769.19. also, we have a a good variance score at 99.3254 %.

**This mean that our deeper and wider Neural Network Model (5 Hidden Layers + 40 Neurons) Variance Score has improved by 3.1% from 96.3320% to 99.3254%.**


#### 7.3.6 Neural Network Actual vs Prediction Visualization<a id='7.3.6_Neural_Network_Actual_vs_Prediction_Visualization'></a>

```python
# Let's plot the predictions against the actual result. Use scatter()
plt.figure(figsize=(15,10))
plt.scatter(y_test,y_pred_nn_1, color='black', marker='D', s=40, alpha=0.3, label='Actual vs Prediction-NN Model 1 (base)')
plt.scatter(y_test,y_pred_nn_2, color='red', marker='o', s=20, alpha=0.3,label='Actual vs Prediction-NN Model 2')
plt.scatter(y_test,y_pred_nn_3, color='blue', marker='x', s=20, alpha=0.3,label='Actual vs Prediction-NN Model 3')
plt.title('NN keras Regression Models: predictions vs. the actual values')
plt.plot([0,(np.max(y_test))],[0,(np.max(y_test))], 'blue', linewidth=30, alpha=0.2)
plt.legend()
plt.xlabel('Actual Result')
plt.ylabel('Prediction')
```
